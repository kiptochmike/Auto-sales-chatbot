```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('Auto Sales data.csv')
df.head(20)

```

```python
def get_order_details(order_number):
    """Get details of a specific order."""
    order = df[df['ORDERNUMBER'] == order_number]
    if order.empty:
        return "Order not found."
    return order.to_dict(orient='records')[0]

def get_total_sales_by_productline(product_line):
    """Get total sales for a specific product line."""
    sales = df[df['PRODUCTLINE'] == product_line]['SALES'].sum()
    return f"Total sales for product line '{product_line}': ${sales:.2f}"

def get_orders_by_date(order_date):
    """Get orders placed on a specific date."""
    orders = df[df['ORDERDATE'] == order_date]
    if orders.empty:
        return f"No orders found for date {order_date}."
    return orders.to_dict(orient='records')

def get_orders_by_customer(customer_name):
    """Get all orders placed by a specific customer."""
    orders = df[df['CUSTOMERNAME'] == customer_name]
    if orders.empty:
        return f"No orders found for customer {customer_name}."
    return orders.to_dict(orient='records')

def get_customer_contact_info(customer_name):
    """Get contact information for a specific customer."""
    customer = df[df['CUSTOMERNAME'] == customer_name]
    if customer.empty:
        return "Customer not found."
    return customer[['PHONE', 'ADDRESSLINE1', 'CITY', 'COUNTRY']].to_dict(orient='records')[0]

def get_orders_by_dealsize(deal_size):
    """Get orders with a specific deal size."""
    orders = df[df['DEALSIZE'] == deal_size]
    if orders.empty:
        return f"No orders found with deal size '{deal_size}'."
    return orders.to_dict(orient='records')
```

```python
print(get_order_details(10107))
print(get_total_sales_by_productline('Motorcycles'))
print(get_orders_by_date('24/02/2018'))
print(get_orders_by_customer('Land of Toys Inc.'))
print(get_customer_contact_info('Land of Toys Inc.'))
print(get_orders_by_dealsize('Small'))
```
